## get data from csv

```python
fileToRead = DEFAULT_PATH + "job2024Clean.csv"
```

```python
import pandas as pd

df = pd.read_csv(fileToRead)
df.head(5)
```

## Check the data

```python
df.groupby('jobType').describe()
```

```python
import matplotlib.pyplot as plt

df['jobType'].value_counts()

# Assuming df is your DataFrame containing 'jobType' column
# Group by 'jobType' and calculate the count of rows for each group
grouped_df = df.groupby('jobType').size()

# Plotting the bar chart
grouped_df.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(12, 6))

# Adding labels and title
plt.xlabel('Job Type')
plt.ylabel('Number of Job Posts')
plt.title('IT Job Recruitment Data (13/11/2023 - 13/12/2023) on Glint, Indeed, CareerBuilder')

# Show the plot
plt.show()
```

## process the data to other csv file

```python

# Assuming df is your DataFrame containing 'jobType' column
# Group by 'jobType' and calculate the count of rows for each group
grouped_df = df[['jobType']].value_counts()
sorted_jobs = grouped_df.sort_values(ascending=False)

# Select only the top 5 job types
top_5_jobs = sorted_jobs.head(5)

# Plotting the bar chart
ax = top_5_jobs.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(12, 6))

# Adding labels and title
plt.xlabel('Job Type')
plt.ylabel('Number of Job Posts')
plt.title('Top 5 IT Job Types (13/11/2023 - 13/12/2023) on Glint, Indeed, CareerBuilder')

# Annotate each bar with the count on top
for i, v in enumerate(top_5_jobs):
    ax.text(i, v + 2, str(v), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python


# Assuming df is your DataFrame containing 'jobType', 'month', and 'year' columns
# Filter the DataFrame for the specified month and year
filtered_df = df[(df['month'] == 1) & (df['year'] == 2024)]

# Group by 'jobType' and calculate the count of rows for each group
grouped_df = filtered_df['jobType'].value_counts()

# Select only the top 5 job types
top_5_jobs = grouped_df.head(5)

# Plotting the bar chart
ax = top_5_jobs.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(12, 6))

# Adding labels and title
plt.xlabel('Job Type')
plt.ylabel('Number of Job Posts')
plt.title('Top 5 IT Job Types (January 2024)')

# Annotate each bar with the count on top
for i, v in enumerate(top_5_jobs):
    ax.text(i, v , str(v), ha='center', va='bottom')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```
