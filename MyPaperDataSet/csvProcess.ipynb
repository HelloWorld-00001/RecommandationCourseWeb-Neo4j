```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile, open(output_file, 'w', newline='') as outfile:
        reader = csv.DictReader(infile)
        fieldnames = reader.fieldnames

        # Ensure that the required columns exist in the CSV file
        required_columns = ['Knowledge', 'Tool', 'Framework', 'Platform', 'Programing Language']
        for col in required_columns:
            if col not in fieldnames:
                print(f"Error: {col} column not found in the CSV file.")
                return

        writer = csv.DictWriter(outfile, fieldnames=fieldnames)
        writer.writeheader()

        for row in reader:
            for col in required_columns:
                if col in row:
                    row[col] = ','.join([value.strip() for value in row[col].split(',')])
                    print(row[col])

            writer.writerow(row)

    print(f"CSV file processed successfully. Output written to {output_file}")

# Replace 'input.csv' and 'output.csv' with your actual file names
process_csv('/Users/nguyenvanviet/Work/Courses/DataGen/MyPaperDataSet/courseFinal.csv', 'output.csv')
```
