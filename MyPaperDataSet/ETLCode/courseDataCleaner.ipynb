```python
JOB_LIST = ["Software Architect", "Business Analyst", "Data Scientist", "Game Development",
            "Database Administrator", "Data Engineer", "Data Analysts","Software Engineer", "Web Development"]
DEFAULT_PATH = "/Users/nguyenvanviet/Work/Courses/DataGen/MyPaperDataSet/"
COMPETENCIES_LIST = ["Knowledge", "Platform", "Framework", "ProgrammingLanguage", "Tool"]
```

```python
import pandas as pd
import sparknlp as sp
import csv
```

## read the raw course data and tranfer it to dataframe

```python
courseFilePath = DEFAULT_PATH + "course_1.csv"
courseData = pd.read_csv(courseFilePath, encoding='latin1')
courseData.groupby('Level').describe()
```

## clear un used columns

```python
courseData = courseData.drop(['Unnamed: 23', 'Unnamed: 24', 'Program', 'RelationInsOrg'], axis=1)
courseData.head(5)
```

## Filter and To int all the numberic columns
#### Handle null and none-numberic values -> convert to numeric

```python
courseData['Price'] = courseData['Price'].fillna('$-1')
courseData[['Rating', 'Enroll']] = courseData[['Rating', 'Enroll']].fillna("0")

courseData['Price'] = courseData['Price'].apply(lambda x: x[1:] if x[0] == '$' else '0')

courseData['Enroll'] = courseData['Enroll'].replace(",","")
courseData[['Price', 'Rating', 'Enroll']] = courseData[['Price', 'Rating', 'Enroll']].astype(float)

courseData['Enroll'] = courseData['Enroll'].astype(int)
courseData.head(5)
```

## Dectect web data from link

```python
url = "https://www.gooogleas.com/course/staadpro-3d-struc"

def getWebInfor(url):
    # Extract webname
    webname_start_index = url.find("www.") + 4
    webname_end_index = url.find(".", webname_start_index)
    webname = url[webname_start_index:webname_end_index]

    # Extract URL
    url_start_index = url.find("://") + 3
    url_end_index = url.find("/", url_start_index)
    webUrl = url[url_start_index:url_end_index]
    return webname, webUrl

courseData[['webname', 'weburl']] = courseData['Link'].apply(lambda x: pd.Series(getWebInfor(x)))
courseData.head(5)
```

## Get the competency data from csv, and detect it form other string

```python
Competencies = {"Knowledge": [], "Platform": [], "Framework":[], "ProgrammingLanguage":[], "Tool":[]}

for comp in COMPETENCIES_LIST:
    temp = pd.read_csv(DEFAULT_PATH + "Competency/" + comp + ".csv")
    Competencies[comp] = temp[comp[:1].lower() + comp[1:]].dropna().to_list()
    print(len(Competencies[comp]))
```

```python

def detectCompetency(df, search_string):
    search_string = search_string.upper()
    matching_values_dict = {}

    for column in COMPETENCIES_LIST:
        matching_values_dict[column] = []
        if len(df[column]) < 1:
            continue

        for value in df[column]:
            if value == '':
                continue

            idx1 = search_string.find(" " + value + " ")
            idx2 = search_string.find(" " + value + ",")
            idx3 = search_string.find("," + value + ",")
            idx4 = search_string.find("," + value + " ")
            idx5 = search_string.find("." + value + " ")
            idx6 = search_string.find("." + value + ",")
            idx7 = search_string.find("." + value + ".")
            idx8 = search_string.find("," + value + ".")
            idx9 = search_string.find(" " + value + ".")
            final = max(idx1, idx2, idx3, idx4, idx5, idx6, idx7, idx8, idx9)

            if final >= 0:
                matching_values_dict[column].append(value)

    return matching_values_dict

```

### detect competency from each skill requirement row of raw data

```python
# Process each row and add competency columns
for index, row in courseData.iterrows():
    skill_requirement = row['skillrequirement']
    if pd.isna(skill_requirement) or len(skill_requirement) < 2:
        for comp in COMPETENCIES_LIST:
            courseData.at[index, "Need"+comp] = pd.NA
        continue    
    
    competency_values = detectCompetency(Competencies, skill_requirement)
        
    #Add competency values as new columns
    for competency, values in competency_values.items():
        courseData.at[index, "Need"+competency] = ','.join(values)
    
courseData = courseData.replace("", pd.NA)
courseData.head(5)
```

## write the clean data to csv

```python
courseData.to_csv(DEFAULT_PATH + "courseClean.csv", index=False)
```
